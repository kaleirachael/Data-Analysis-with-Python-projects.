```python
import numpy as np

def calculate(list):
  if not len(list) == 9:
    raise ValueError('List must contain nine numbers.')


  list = np.array(list)
  list = np.array(list).reshape(3,3)

  mean = (list.mean(axis=0), list.mean(axis=1), list.mean())
  mean = [array.tolist() for array in mean]

  var = (list.var(axis=0), list.var(axis=1), list.var())
  var = [array.tolist() for array in var]

  std = (list.std(axis=0), list.std(axis=1), list.std())
  std = [array.tolist() for array in std]

  max = (list.max(axis=0), list.max(axis=1), list.max())
  max = [array.tolist() for array in max]

  min = (list.min(axis=0), list.min(axis=1), list.min())
  min = [array.tolist() for array in min]
  
  sum = (list.sum(axis=0), list.sum(axis=1), list.sum())
  sum = [array.tolist() for array in sum]

  values = dict()
  values['mean'] = mean
  values['variance'] = var
  values['standard deviation'] = std
  values['max'] = max
  values['min'] = min
  values['sum'] = sum

  return(values)
```

```python
calculate([0,1,2,3,4,5,6,7,8])
```

```python
calculate([9,1,5,3,3,3,2,9,0])
```
